```python
import os
import random
import sys

import cv2
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import seaborn as sns
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Lambda, ELU, Activation
from keras.layers.convolutional import Convolution2D
from tensorflow.keras.layers import BatchNormalization

from game.wrapped_flappy_bird import GameState

gamma = 0.99
final_epsilon = 0.001
initial_epsilon = 1

number_of_iterations = 500000
replay_memory_size = 1000
minibatch_size = 32
size1 = (84, 84)

def imagepr(frame):
    frame = frame[50:250,0:400]

    frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    frame = cv2.resize(frame,size1,interpolation=cv2.INTER_AREA)
    _ , frame = cv2.threshold(frame,1,255,cv2.THRESH_BINARY)

    frame = frame.astype(np.float32)/255

    return  frame

def train(model):
    game_state = GameState()

    replay_memory = []

    action = torch.zeros([2], dtype=torch.float32)
    action[0] = 1
    
    image_data, reward, terminal = game_state.frame_step(action)
    image_data = imagepr(image_data)
    
    state = np.stack((image_data, image_data, image_data, image_data), axis=2)

    epsilon = initial_epsilon
    iteration = 0

    epsilon_decrements = np.linspace(initial_epsilon, final_epsilon, number_of_iterations*2//3)
    epsilon_decrements = np.append(epsilon_decrements, np.array([final_epsilon]*(number_of_iterations//3+2)))

    while iteration < number_of_iterations:

        output = model(np.expand_dims(state, axis=0))

        action = np.zeros(2)

        random_action = random.random()
        if random_action < epsilon:
            if random.random() < 0.1:
                action_index = 1
            else:
                action_index = 0
        else:
            action_index = np.argmax(output)
        action[action_index] = 1

        image_data_1, reward, terminal = game_state.frame_step(action)
        image_data_1 = imagepr(image_data_1)
        image_data_1 = np.reshape(image_data_1, (84, 84, 1))

        state_1 = np.append(image_data_1, state[:, :, :3], axis=2)

        replay_memory.append([state, action, reward, state_1, terminal])

        if len(replay_memory) > replay_memory_size:
            replay_memory.pop(0)

        epsilon = epsilon_decrements[iteration]

        if not iteration%10:
            
            minibatch = np.array(random.sample(replay_memory, min(len(replay_memory), minibatch_size)))
            state_batch = np.array([elem for elem in minibatch[:, 0]])
            action_batch = np.array([1 if elem[0]==0 else 0 for elem in minibatch[:, 1]])
            reward_batch = minibatch[:, 2]
            state_1_batch = np.array([elem for elem in minibatch[:, 3]])
            terminal_batch = minibatch[:, 4]
            current_prediction_batch = np.array(model.predict(state_batch, verbose = 0))
            next_prediction_batch = np.array(model.predict(state_1_batch, verbose = 0))

            real = reward_batch + gamma * next_prediction_batch.max(1) * (1-terminal_batch)
            current_prediction_batch[range(minibatch.shape[0]), action_batch.ravel()] = real.ravel()

            loss = model.train_on_batch(state_batch, current_prediction_batch)

        state = state_1
        iteration += 1

        if iteration % 1 == 0:
            print(iteration, ": ", "e:", round(epsilon, 3), "action:", action_index, "reward:", reward, "q:", np.max(output))
```

```python
def create():
    model = Sequential()
    model.add(Convolution2D(32, (8, 8), strides=4, input_shape=(84, 84, 4)))
    model.add(Activation('relu'))
    model.add(Convolution2D(64, (4, 4), strides=2))
    model.add(Activation('relu'))
    model.add(Convolution2D(64, (3, 3), strides=1))
    model.add(Activation('relu'))
    
    model.add(Flatten())
    model.add(Activation('relu'))
    model.add(Dense(512))
    model.add(Activation('relu'))
    model.add(Dense(2))

    model.compile(optimizer="adam", loss="mse")

    return model
```

```python
model = create()
train(model)
```
