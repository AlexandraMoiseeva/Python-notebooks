```python
import numpy as np
from numpy import linalg
from matplotlib import pyplot as plt  
```

```python
tau = 0.05
h = 2.0
m = 10 #ny
l = 10 #nx
n = 10 #tmax
y0=1
x0=1
y1 = 3
x1 = 2
m1 = 3
l1 = 5
t = 0
a = 34
w = np.pi/4
```

```python
k = (a*tau/h)**2
```

```python
n1 = np.zeros((m+1, l+1))
n2 = np.zeros((m+1, l+1))
v = np.zeros(n+1)
```

```python
for i in range(1, n+1):
    n0 = np.zeros((m+1, l+1))
    n0 = n0 - n1
    n0 = n0 + n2*(2-4*k)
    n0 = n0 + (np.roll(n2, 1) + np.roll(n2, -1) + np.roll(n2.T, 1).T + np.roll(n2.T, -1).T)*k
    if (i>=t):
        n0[y0][x0] = n0[y0][x0] + np.sin(w*(i-t)*tau)
    n0[(m+1-m1)//2:(m+1+m1)//2, (l+1-l1)//2:(l+1+l1)//2] = np.zeros((m1, l1))
    v[i] = n0[y1][x1]
    n1 = n2
    n2 = n0
```

```python
plt.plot(v)
```
