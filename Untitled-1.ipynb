```python
from tkinter import *
from  tkinter import ttk
import random

ws  = Tk()
ws.title('Табель учета рабочего времени')
ws.geometry('1280x720')
ws['bg'] = '#AC99F2'

global k
k = 0
def input_record():
    
    global count 
    
    my_game.insert(parent='',index='end',iid = count,text='',values=(playerid_entry.get(),playername_entry.get(),playerrank_entry.get()))
    count += 1

   
    playerid_entry.delete(0,END)
    playername_entry.delete(0,END)
    playerrank_entry.delete(0,END)
    
#save Record
def update_record():
    selected=my_game.focus()
    #save new data 
    my_game.item(selected,text="",values=(playerid_entry.get(),playername_entry.get(),playerrank_entry.get()))
    
   #clear entry boxes
    playerid_entry.delete(0,END)
    playername_entry.delete(0,END)
    playerrank_entry.delete(0,END)
    
#Select Record
def select_record():
    #clear entry boxes
    playerid_entry.delete(0,END)
    playername_entry.delete(0,END)
    playerrank_entry.delete(0,END)
    
    #grab record
    selected=my_game.focus()
    #grab record values
    values = my_game.item(selected,'values')
    #temp_label.config(text=selected)

    #output to entry boxes
    playerid_entry.insert(0,values[0])
    playername_entry.insert(0,values[1])
    playerrank_entry.insert(0,values[2])
```
